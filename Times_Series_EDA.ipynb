## Time Resampler

```python
#there are multiple rule of Resampler
df_bmw.resample(rule = 'A').min()
```

```python
#this is giving you each year lastday value
```

```python
df_bmw.resample(rule='A').max()
```

```python
df_bmw.resample(rule = 'A').max()['Open'].plot()
```

```python
#Quaterly startfrequency

df_bmw.resample(rule = 'QS').max()['High'].plot()
```

```python
df_bmw['Open'].resample(rule = 'A').mean().plot(kind = 'bar')
```

#Rolling

```python
df_bmw['High'].rolling(11).max().head(20)
```

```python
df_bmw['Open:30 days rolling'] = df_bmw['Open'].rolling(30).mean()
```

```python
df_bmw.head(31)
```

```python
df_bmw[['Open','Open:30 days rolling']].plot(figsize = (12,5))
```
